```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
plt.rc('axes', unicode_minus=False)
plt.rc('font', family='Malgun Gothic')
%matplotlib inline
```

```python
from IPython.display import set_matplotlib_formats
# 폰트가 선명하게 보이기 위해
set_matplotlib_formats('retina')
```

```python
df = pd.read_csv("./data/people_hospital2.csv", low_memory=False)
```

```python
df
```

```python
fig, ax1 = plt.subplots(figsize=(20, 8))
plt.title("행정동 노인인구별 병원수")

ax1.set_xlabel("행정동 노인인구")
ax1.set_ylabel("병원수")
ax1.bar(df["행정동명"], df["인구수"], color="c", label="노인인구")
ax1.legend(loc=2)

ax2= ax1.twinx()
ax2.set_ylabel("병원수")
ax2.plot(df["행정동명"], df["병원"], color="m", linewidth=4.0, label="병원수")
ax2.legend(loc=1)

plt.legend()
plt.grid()

plt.show()
```

```python
df.corr()
```

```python
# lmplot 시각화 (회귀선)
sns.lmplot(data = df, x = '인구수', y = '병원') 
plt.title("노인인구수와 병원수의 상관관계")
```
