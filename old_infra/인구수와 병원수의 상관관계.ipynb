```python
df = pd.read_csv("./data/people_hospital2.csv", low_memory=False)
df2 = pd.read_csv("./data/people_hospital3.csv", low_memory=False)
```

```python
df
df2
```

```python
# 인구수 기준으로
fig, ax1 = plt.subplots(figsize=(20, 8))
plt.title("행정동 노인인구별 병원수")

ax1.set_xlabel("행정동 노인인구")
ax1.set_ylabel("병원수")
ax1.bar(df["행정동명"], df["인구수"], color="c", label="노인인구")
ax1.legend(loc=2)

ax2= ax1.twinx()
ax2.set_ylabel("병원수")
ax2.plot(df["행정동명"], df["병원"], color="m", linewidth=4.0, label="병원수")
ax2.legend(loc=1)

plt.legend()
plt.grid()

plt.show()
```

```python
# 병원수 기준으로
fig, ax1 = plt.subplots(figsize=(20, 8))
plt.title("행정동 노인인구별 병원수")

ax1.set_xlabel("행정동 노인인구")
ax1.set_ylabel("병원수")
ax1.bar(df2["행정동명"], df2["인구수"], color="c", label="노인인구")
ax1.legend(loc=2)

ax2= ax1.twinx()
ax2.set_ylabel("병원수")
ax2.plot(df2["행정동명"], df2["병원"], color="m", linewidth=4.0, label="병원수")
ax2.legend(loc=1)

plt.legend()
plt.grid()

plt.show()
```
